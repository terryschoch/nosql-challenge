```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

Unless otherwise stated, for each question:

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query_hyg = {'scores.Hygiene': {'$eq': 20}}    
results_hyg = establishments.find(query_hyg)

# Use count_documents to display the number of documents in the result
print(f'Number of documents with a Hygiene score of 20: {establishments.count_documents(query_hyg)}')

# Display the first document in the results using pprint
for i in range(1):
    pprint(results_hyg[i])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(results_hyg)

# Display the number of rows in the DataFrame
print(f'{len(hygiene_df.index)}')

# Display the first 10 rows of the DataFrame
hygiene_df
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query_rval = {'LocalAuthorityName': {'$regex': 'London'}, 
         'RatingValue': {'$gte': '4'}}   
results_rval = establishments.find(query_rval)

# Use count_documents to display the number of documents in the result
print(f'Number of documents with a Rating Value of 4 or more: {establishments.count_documents(query_rval)}')

# Display the first document in the results using pprint
for i in range(1):
    pprint(results_rval[i])
```

```python
# Convert the result to a Pandas DataFrame
rvalue_df = pd.DataFrame(results_rval)

# Display the number of rows in the DataFrame
print(f'{len(hygiene_df.index)}')

# Display the first 10 rows of the DataFrame
rvalue_df
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

query_near = {'geocode.latitude': {'$gte': 51.48014200},
              'geocode.latitude': {'$lte': 51.50014200},
              'geocode.longitude': {'$gte': 0.07384000},
              'geocode.longitude': {'$lte': 0.09384000},
#               'geocode.latitude': {'$gte': latitude + degree_search},
#               'geocode.latitude': {'$lte': latitude - degree_search},
#               'geocode.longitude': {'$gte': longitude - degree_search},
#               'geocode.longitude': {'$lte': longitude + degree_search},
              'RatingValue': {'$eq': '5'}}   

# Sort in ascending order by Hygiene score
sort = [('scores.Hygiene', 1)]

# Limit the results to the top 5
limit = 5

results_near = (list(establishments.find(query_near).sort(sort).limit(limit)))

# Print the results
pprint(list(establishments.find(query_near).sort(sort).limit(limit)))
```

```python
# Convert result to Pandas DataFrame
near_df = pd.DataFrame(results_near)
near_df
```
